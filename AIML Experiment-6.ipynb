```python
#Experiment-6 of AIML

import pandas as pd
import numpy as np

#create a sample dataset
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [25, 30, 35, 40, 45],
    'Gender': ['F', 'M', 'M', 'M', 'F'],
    'Marks': [85, 90, 78, 92,100], # Added 'Marks' column
    'Address': ['123 Main St', '456 Oak Ave', '789 Pine Ln', '101 Elm St', '222 Maple Rd'], # Adjusted length to match other columns
    'City': ['Los Angeles', 'New York', 'Houston', 'Los Angeles', 'Miami'], # Adjusted length to match other columns
    'Rank': [2, 1, 4, 3, 8], # Adjusted length to match other columns
    'Grade': ['B', 'A', 'C', 'B', 'D'] # Adjusted length to match other columns
}

#create a pandas dataframe from the dataset

df = pd.DataFrame(data)
print("Sample DataFrame:")

print(df)

df.head(4)  #Printed 4 records from the dataset and generated the related graphs

df.info()   #Getting information on the dataset content

df.isna().sum()

#listwise deletion
df_listwise = df.dropna()
print("\nDataFrame after listwise deletion:")
print(df_listwise)

# Mean Imputation
df_mean_imputed = df.copy()
mean_marks = df_mean_imputed['Marks'].mean()
mean_marks = df_mean_imputed['Marks'].mean()
df_mean_imputed['Marks'].fillna(mean_marks, inplace=True)

print("\nDataFrame after mean imputation:")
print(df_mean_imputed)
df_mean_imputed['Marks'].fillna(mean_marks, inplace=True)

print("\nDataFrame after mean imputation:")
print(df_mean_imputed)


#Outliers detection method:-

from scipy import stats

#calculate z-scores
z_scores = np.abs(stats.zscore(df['Marks']))
threshold = 80 #define a threshold for Z-score
outliers = np.where(z_scores > threshold)

print("\nOutliers based on Z-score:")
print(df.iloc[outliers])


outliers

#IOR method
Q1 = df['Marks'].quantile(0.25)
Q3 = df['Marks'].quantile(0.75)
IQR = Q3 - Q1

# define bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# identify outliers
outliers = df[(df['Marks'] < lower_bound) | (df['Marks'] > upper_bound)]

# remove outliers
df_cleaned = df[(df['Marks'] >= lower_bound) & (df['Marks'] <= upper_bound)]

print("\nDataFrame after removing outliers (IQR method):")
print(df_cleaned)
```
